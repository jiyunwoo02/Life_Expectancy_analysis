### 데이터 불러오기 
결측치 처리된 데이터 불러오기

```python
import pandas as pd

data = pd.read_csv('Life Expectancy Data.csv')
data.head()
```

### 컬럼 공백 제거

```python
original_cols = list(data.columns)
new_cols = []

# 컬럼 공백 제거 및 소문자 변형
for col in original_cols:
    new_cols.append(col.strip().replace('  ', ' '))
data.columns = new_cols

data.head()
```

### 레이블 인코딩

```python
data.info()
```

```python
data.Status = data.Status.replace('Developing', 0)
data.Status = data.Status.replace('Developed', 1)
data.head()
```

### 나라 및 B형간염 컬럼 삭제

```python
data = data.drop(labels='Country', axis=1)
data = data.drop(labels='Hepatitis B', axis=1)
data.head()
```

### 남은 결측치 제거

```python
data = data.dropna() #결측치제거
data.info()
```

### 컬럼별 정규화 진행

```python
# 정규화하기
def normalize( x ) :
    return ( x - min(x) ) / ( max(x) - min(x) )

for i in range( len(data.columns) -2 ) :
    col = data.columns[i]
    data[col] = normalize( data[col] )
    
data.head()
```

### 분포도
기대수명 분포도 확인하기 

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(data['Life expectancy'])
```

### 상관관계
상관관계 매트릭스 + 하나의 그래프

```python
d = data.corr()

sns.heatmap(d, 
               annot = True,    
               cmap = 'RdYlBu_r', 
               vmin = -1, vmax = 1,
              )
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(d[['Life expectancy']], annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation with Life Expectancy')
plt.show()
```

### 스캐터플랏
상관관계를 기반으로 한 스캐터플랏 그리기

```python
y = data['Life expectancy']

# 0.5이상의 상관성을 가진 컬럼들
x1 = data['Schooling'] #0.78
x2 = data['Income composition of resources'] #0.76
x3 = data['BMI'] #0.6
x4 = data['Status'] #0.51
x5 = data['HIV/AIDS'] #-0.57
x6 = data['Adult Mortality'] #-0.67

x = [x1, x2, x3, x4, x5, x6]
```

```python
import matplotlib.pyplot as plt

for i in range(6):
    plt.title('scatter graph x%d'%(i+1))
    plt.scatter(x[i], y, c=data['Status']) #나라기반으로 색상구분하기!
    plt.colorbar()
    plt.show()
```
