```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Life Expectancy Data.csv')
df
```

```python
df.rename(columns={' thinness  1-19 years':'thinness 10-19 years'}, inplace=True) # 컬럼명 수정
```

```python
original_cols = list(df.columns)
new_cols = []

# 컬럼 공백 제거 및 소문자 변형
for col in original_cols:
    new_cols.append(col.strip().replace('  ', ' ').lower())
df.columns = new_cols
```

```python
# Checking for Null Values
df.isnull().sum()
```

```python
# 'life expectancy' 컬럼의 값이 결측치인 행 추출
missing_life_expectancy_rows = df[df['life expectancy'].isna()]

# 결과 확인
missing_life_expectancy_rows
```

1) life expectancy 컬럼의 값이 결측치인 행 추출해 관찰한 결과: 10개
   
→ life expectancy의 값이 NaN인 행 
   
- 모두 adult mortality 값 모두 NaN
- alcohol 하나 NaN
- gdp 5개 NaN
- population 8개 NaN
- thinness 5-9랑 10-19 2개 NaN (서로 동일한 행)
- 모두 developing
- 등등

```python
missing_adult_mortality_rows = df[df['adult mortality'].isna()]

# 결과 확인
missing_adult_mortality_rows
```

2) adult mortality 컬럼의 값이 결측치인 행 추출해 관찰한 결과: 10개
   
→ 1)과 동일하다

```python
missing_diphtheria_rows = df[df['diphtheria'].isna()]

# 결과 확인
missing_diphtheria_rows
```

3) diphtheria 컬럼의 값이 결측치인 행 추출해 관찰한 결과: 19개
   
→ diphtheria의 값이 NaN인 행 
   
- polio 모두 NaN

- 등등

```python
missing_bmi_rows = df[df['bmi'].isna()]

# 결과 확인
missing_bmi_rows
```

4) bmi 컬럼의 값이 결측치인 행 추출해 관찰한 결과: 34개
   
→ bmi의 값이 NaN인 행
   
- 모두 developing

- life expectancy-adult mortality 2개 NaN
- 등등

```python
# 'country' 컬럼을 기준으로 데이터 그룹화
grouped_by_country = df.groupby('country')

# 각 나라별로 데이터 출력 (예시로 5개씩 출력)
for country, group in grouped_by_country:
    print(f"Country: {country}")
    print(group.head(5))  # 각 나라별로 5개의 행만 출력하도록 설정
    print("\n")
```
